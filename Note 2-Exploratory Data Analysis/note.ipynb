## IPOs data from Web

```python
import pandas as pd
import requests

# Set headers to mimic a real browser request
# This prevents the website from blocking our script as a bot
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
}

# URL for the 2023 IPO listings page on Stock Analysis website
url = "https://stockanalysis.com/ipos/2023/"

# Send HTTP GET request to fetch the webpage content
response = requests.get(url, headers=headers)

# Parse all HTML tables from the page into a list of DataFrames
# Each table on the page becomes a separate DataFrame in the list
ipo_dfs = pd.read_html(response.text)

# Access the first table with: ipo_dfs[0]
```

```python
# "stacking dataframes"
# pd.concat() is used to concatenate the DataFrames vertically.
# The ignore_index=True parameter ensures that the resulting DataFrame has a new index,
# ignoring the original indices of the input DataFrames.
# The stacked_df now contains the concatenated DataFrame.
stacked_ipos_df = pd.concat([ipos_2024, ipos_2023], ignore_index=True)
```

```python
# save the stacked dataframe to a csv file
stacked_ipos_df.to_csv('stacked_ipos_2023_2024.csv', index=False)
```
